<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<h2 id="Satellite-data-sources" style="direction: rtl; unicode-bidi: plaintext; text-align: right;">مصادر بيانات الأقمار الصناعية</h2>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
يوفّر أسطول
<span dir="ltr" style="unicode-bidi: isolate;">NASA</span>
من الأقمار الصناعية المخصّصة لرصد الأرض مصادر عديدة من المعلومات حول نظام مناخ الأرض.
يعرض موقع
<a href="https://eyes.nasa.gov/apps/earth/" style="direction: rtl; unicode-bidi: plaintext;">
<span dir="ltr" style="unicode-bidi: isolate;">Eyes on the Earth</span>
</a>
مواقع الكثير من هذه الأقمار الصناعية في الزمن الحقيقي.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
تمتلك معظم أقمار
<span dir="ltr" style="unicode-bidi: isolate;">NASA</span>
المدارية الخاصة برصد الأرض مدارات قطبية متزامنة مع الشمس
(<strong>Sun-synchronous, polar orbits</strong>).
وهذا يعني أن القمر الصناعي يمرّ فوق الموقع نفسه تقريبًا في الوقت المحلي نفسه من اليوم،
مما يجعل المقارنات عبر الزمن أكثر اتساقًا (مثل مقارنة صور الغطاء النباتي يومًا بعد يوم أو سنة بعد سنة).
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
كما أن كثيرًا من هذه الأقمار الصناعية تعبر خط الاستواء في وقت محلي ثابت تقريبًا
(مثل “مرور صباحي” أو “مرور بعد الظهر”).
وهذه الخاصية مفيدة لأن الإضاءة والظلال وزاوية الشمس يمكن أن تؤثر في بعض القياسات.
</p>
</div>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<h2 id="Multi-sensor-datasets" style="direction: rtl; unicode-bidi: plaintext; text-align: right;">مجموعات بيانات متعددة المستشعرات</h2>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
تقيس بعثات أقمار صناعية مختلفة—سواء كانت متداخلة زمنيًا أو تفصل بينها سنوات—أحيانًا المتغيّر نفسه.
ويُعد <strong>الهطول المطري</strong> أحد أهم المتغيّرات المناخية للزراعة،
وقد تم قياسه عبر عدة بعثات ضمن كوكبة
<span dir="ltr" style="unicode-bidi: isolate;">NASA</span>
لـ
<span dir="ltr" style="unicode-bidi: isolate;">Global Precipitation Measurement (GPM)</span>.
</p>

<img src="./assets/GPM-constellation.jpg" alt="GPM constellation">
</div>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<h3 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">بيانات الهطول من&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">IMERG</span></h3>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
تجمع خوارزمية
<span dir="ltr" style="unicode-bidi: isolate;">Integrated Multi-satellitE Retrievals for GPM (IMERG)</span>
بيانات هذه المهام المختلفة لتقدير إجمالي الهطول المطري على مستوى العالم
بفواصل زمنية قدرها 30 دقيقة وبدقة مكانية تقارب 10 كم.
توجد ثلاثة منتجات رئيسية من
<span dir="ltr" style="unicode-bidi: isolate;">IMERG</span>
تختلف في طريقة دمج البيانات وفي “الكمون” (أي مدى سرعة إتاحتها بعد الرصد):
</p>

<ul style="direction: rtl; unicode-bidi: plaintext; text-align: right; padding-right: 1.5em;">
  <li><strong>IMERG Early</strong>: متاح بسرعة أكبر (زمن تأخير أقل) لكنه عادةً أقل دقة من المنتج النهائي.</li>
  <li><strong>IMERG Late</strong>: متاح بعد وقت أطول قليلًا ويوازن بين السرعة والدقة.</li>
  <li><strong>IMERG Final</strong>: المنتج الأكثر اكتمالًا/دقة، لكنه يأتي بزمن تأخير أكبر لأنه يتضمن معايرة إضافية.</li>
</ul>
</div>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<h2 id="Using-satellite-based-precipitation-data" style="direction: rtl; unicode-bidi: plaintext; text-align: right;">استخدام بيانات الهطول المعتمدة على الأقمار الصناعية</h2>
</div>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
شهدت أجزاء من شمال الجزائر وتونس فيضانات خاطفة في مايو 2023.
سنستخدم منتج
<span dir="ltr" style="unicode-bidi: isolate;">IMERG-Late</span>
لتقدير إجمالي الهطول الذي سقط خلال يوم واحد في تلك المنطقة.
</p>
</div>

```python
import earthaccess
import xarray as xr
from matplotlib import pyplot

auth = earthaccess.login()
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
على الرغم من أن
<span dir="ltr" style="unicode-bidi: isolate;">IMERG-Final</span>
يأتي بفواصل زمنية قدرها 30 دقيقة،
إلا أننا سنستخدم اليوم
<a href="https://disc.gsfc.nasa.gov/datasets/GPM_3IMERGDF_06/summary" style="direction: rtl; unicode-bidi: plaintext;">
نسخة من البيانات تم تجميعها إلى خطوات زمنية يومية
</a>.
</p>
</div>

```python
results = earthaccess.search_data(
    short_name = 'GPM_3IMERGDF',
    temporal = ('2023-05-25', '2023-05-30'))

results[0]
```

```python
# NOTE: open() requires a sequence of file references
files = earthaccess.open(results)
files
```

```python
ds = xr.open_dataset(files[1])
ds
```

```python
list(ds.variables.keys())
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
تحتوي مجموعة البيانات هذه على متغيّرات عديدة، وكلها موصوفة في
<a href="https://disc.gsfc.nasa.gov/datasets/GPM_3IMERGDF_06/summary" style="direction: rtl; unicode-bidi: plaintext;">
وثائق&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">IMERG-Late</span>
</a>.
سنستخدم متغيّر
<span dir="ltr" style="unicode-bidi: isolate;">precipitation</span>،
وهو تقدير الهطول المُعايَر بالمقاييس المطرية والمستند إلى عدة أقمار صناعية،
والذي توصي به
<span dir="ltr" style="unicode-bidi: isolate;">NASA</span>
للاستخدام العام.
</p>
</div>

```python
ds['precipitation']
```

```python
# NOTE: vmax = 100 makes it easier to see lower precipitation values
ds['precipitation'].plot(vmax = 100)
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
يبدو هذا غريبًا! هل لديك فكرة عمّا الخطأ؟
</p>

<h4 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">&#x1F6A9; <span style="color:red">انتبه</span></h4>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
نحتاج إلى تدوير الشكل بحيث تقابل صفوف الصورة نطاقات خطوط العرض.
<strong>تحديدًا، نحتاج إلى إخبار&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">xarray</span>&nbsp;
بأن خط الطول (<span dir="ltr" style="unicode-bidi: isolate;">lon</span>) يجب أن يمتد على محور&nbsp;<span dir="ltr">X</span>
وأن خط العرض (<span dir="ltr" style="unicode-bidi: isolate;">lat</span>) يجب أن يمتد على محور&nbsp;<span dir="ltr">Y</span>.
</strong>
</p>
</div>

```python
ds['precipitation'].plot(x = 'lon', y = 'lat', vmax = 100)
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
هذا أفضل. لكن لماذا يبدو معظم الشكل داكنًا؟
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
أسهل طريقة لرؤية تفاصيل أكثر هي أن نطلب من
<span dir="ltr" style="unicode-bidi: isolate;">xarray</span>
تمديد شريط الألوان بحيث لا تهيمن القيم المتطرفة على المجال اللوني؛
نقوم بذلك باستخدام
<span dir="ltr" style="unicode-bidi: isolate;">robust = True</span>.
</p>
</div>

```python
ds['precipitation'].plot(x = 'lon', y = 'lat', robust = True)
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
بدلًا من ذلك، يمكننا تحديد القيمة العظمى التي تُسند إلى لون ما باستخدام وسيط الكلمة المفتاحية
<span dir="ltr" style="unicode-bidi: isolate;">vmax</span>.
يوجد وسيط مقابل للقيمة الصغرى وهو
<span dir="ltr" style="unicode-bidi: isolate;">vmin</span>.
</p>
</div>

```python
ds['precipitation'].plot(x = 'lon', y = 'lat', vmax = 15)
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<h2 id="Zooming-in-on-a-study-area" style="direction: rtl; unicode-bidi: plaintext; text-align: right;">التكبير على منطقة دراسة</h2>
</div>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
كيف يمكننا استخدام هذه البيانات في تطبيقات محلية؟
نحتاج إلى طريقة لتركيز الخريطة على منطقة أصغر.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
يمكن استخدام دالة
<span dir="ltr" style="unicode-bidi: isolate;">slice()</span>
المضمّنة في
<span dir="ltr" style="unicode-bidi: isolate;">Python</span>
مع الدالة
<span dir="ltr" style="unicode-bidi: isolate;">sel()</span>
على كائن
<span dir="ltr" style="unicode-bidi: isolate;">xarray DataArray</span>
لاقتطاع مصفوفة كبيرة إلى مصفوفة أصغر.
هنا سنركّز على صندوق حدودي مستطيل صغير يضم مدينة الجزائر.
</p>
</div>

```python
# Area between 30-40 degree N latitude and between 6 W and 6 E longitude
precip = ds['precipitation'].sel(lat = slice(30, 40), lon = slice(-6, 6))

precip.plot(x = 'lon', y = 'lat')
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<h2 id="Using-cartopy" style="direction: rtl; unicode-bidi: plaintext; text-align: right;">استخدام&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">cartopy</span></h2>
</div>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
من الجيد التحقق من أننا نرسم الجزء الصحيح من العالم، خاصةً لأن بيانات
<span dir="ltr" style="unicode-bidi: isolate;">IMERG</span>
قد تكون “مُدارة” في العرض.
مكتبة
<span dir="ltr" style="unicode-bidi: isolate;">cartopy</span>
هي مكتبة في
<span dir="ltr" style="unicode-bidi: isolate;">Python</span>
توفّر أدوات إضافية لرسم الخرائط.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
في المثال أدناه، تم تغيير خريطة الألوان
<span dir="ltr" style="unicode-bidi: isolate;">cmap</span>
لتسهيل رؤية السواحل الداكنة فوق بيانات الهطول.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
ملاحظة: مصطلح
<span dir="ltr" style="unicode-bidi: isolate;">Plate-Carree</span>
هو اسم شائع/تقني لنظام إحداثيات أسطواني متساوي المستطيلات
(<span dir="ltr" style="unicode-bidi: isolate;">equirectangular</span>).
</p>
</div>

```python
from cartopy import crs

proj = crs.PlateCarree()
style = {
    'projection': proj
}

plot = precip.plot(subplot_kws = style, transform = proj, cmap = 'magma_r', x = 'lon', y = 'lat')
plot.axes.coastlines()
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
يمكنك الاطلاع على خرائط الألوان المتاحة هنا:
<a href="https://matplotlib.org/stable/gallery/color/colormap_reference.html" style="direction: rtl; unicode-bidi: plaintext;">
مرجع خرائط الألوان في&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">matplotlib</span>
</a>.
يمكن عكس أي خريطة ألوان بإضافة
<span dir="ltr" style="unicode-bidi: isolate;">'_r'</span>
إلى نهاية اسم خريطة الألوان.
</p>

<h4 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">&#x1F3AF; أفضل ممارسة</h4>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
<strong>من المهم اختيار خريطة الألوان المناسبة لبياناتك.</strong>
يعتقد البعض أن مقياس ألوان قوس قزح أفضل لأنه يحتوي على “ألوان أكثر”.
لكن خرائط ألوان قوس قزح قد تخلق انطباعات بصرية مضلّلة (مثل إبراز فروق ليست حقيقية)
كما أنها قد تُخفي أنماطًا مهمة في البيانات.
</p>
</div>

```python
plot = precip.plot(subplot_kws = style, transform = proj, cmap = 'jet', x = 'lon', y = 'lat')
plot.axes.coastlines()
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
<strong>
كما أن مقاييس ألوان قوس قزح قد تكون إشكالية لمستخدمي عمى الألوان،
كما يظهر في المحاكاة أدناه لتجربة بعض المشاهدين عند النظر إلى مقاييس ألوان مختلفة.
</strong>
</p>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
<strong>
لحسن الحظ، فإن معظم خرائط الألوان المتاحة في
<span dir="ltr" style="unicode-bidi: isolate;">xarray</span>
و
<span dir="ltr" style="unicode-bidi: isolate;">matplotlib</span>
هي خرائط “خطية إدراكيًا” (Perceptually linear)،
وكثيرٌ منها مناسب أيضًا لمستخدمي عمى الألوان.
</strong>
يمكنك استعراض بعض هذه الخرائط بتفصيل أكبر على موقع الدكتورة
<a href="https://colorbrewer2.org/#type=sequential&scheme=BuGn&n=3" style="direction: rtl; unicode-bidi: plaintext;">
<span dir="ltr" style="unicode-bidi: isolate;">Cynthia Brewer</span>
</a>.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
<strong>إذًا، كم كانت كمية المطر حول مدينة الجزائر في ذلك اليوم؟</strong>
</p>
</div>

```python
precip.sel(lon = 3.059, lat = 36.754, method = 'nearest').values
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
هذه كمية معتبرة من المطر لهذه المنطقة خلال يوم واحد،
لكن من الخريطة أعلاه يتضح أن هناك مناطق قرب الجزائر تلقت مطرًا أكثر.
ما هو أقصى إجمالي للهطول على ساحل الجزائر في ذلك اليوم؟
</p>
</div>

```python
precip.sel(lon = slice(3, 3.2), lat = slice(36.5, 36.8)).max()
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
تذكير سريع: بما أن
<span dir="ltr" style="unicode-bidi: isolate;">values</span>
تُعاد كمصفوفة
<span dir="ltr" style="unicode-bidi: isolate;">NumPy</span>،
يمكننا إجراء العمليات الحسابية على هذه المصفوفات كما لو كانت أعدادًا.
لذلك فإن التحويل من مليمتر إلى متر سهل:
</p>
</div>

```python
precip.sel(lon = slice(3, 3.2), lat = slice(36.5, 36.8)).max() / 1000
```
