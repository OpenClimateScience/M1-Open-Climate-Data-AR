<h2 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
مقدّمة إلى&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">xarray</span>
</h2>
</div>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
سنستخدم حزمة برمجية في&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">Python</span>&nbsp;تُسمّى
<span dir="ltr" style="unicode-bidi: isolate;">xarray</span>
لفتح ملف&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">netCDF4</span>&nbsp;الذي قمنا بتنزيله.
وقد صُمِّمت&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">xarray</span>&nbsp;للعمل مع مجموعات البيانات الشبكية متعدّدة الأبعاد.
</p>

<img src="./assets/xarray-dataset.png" alt="xarray dataset structure illustration">
</div>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
كما هو موضّح في الشكل أعلاه، يمكن لكائن واحد من نوع
<span dir="ltr" style="unicode-bidi: isolate;">xarray.Dataset</span>
أن يحتوي على عدة متغيّرات مثل درجة الحرارة، الهطول المطري، خط العرض، وخط الطول.
يتم تخزين كل متغيّر على شكل مصفوفة، وتحديدًا ككائن من نوع
<span dir="ltr" style="unicode-bidi: isolate;">xarray.DataArray</span>.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
وبينما تبقى متغيّرات خط العرض وخط الطول ثابتة عبر الزمن، ولذلك تُمثَّل كمصفوفات ثنائية الأبعاد
(<span dir="ltr" style="unicode-bidi: isolate;">2D</span>)،
فإن متغيّرات مثل درجة الحرارة والهطول المطري تتغيّر عبر الزمن والمكان معًا،
وبالتالي يمكن تمثيلها كمكعّبات بيانات ثلاثية الأبعاد (<strong>Data Cubes</strong>).
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
يمكن استخدام المحاور&nbsp;<strong>x</strong>&nbsp;و<strong>y</strong>&nbsp;ومحور الزمن&nbsp;<strong>(t)</strong>،
والتي تُسمّى أيضًا <strong>الأبعاد (Dimensions)</strong>،
لاقتطاع المصفوفات حسب فترات زمنية أو مناطق دراسية معيّنة.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
<strong>
عادةً ما نقوم باستيراد&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">xarray</span>&nbsp;باستخدام اسم مختصر لتسهيل الاستخدام.
وفي المثال أدناه، نقوم أيضًا باستيراد الوحدة&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">pyplot</span>&nbsp;من مكتبة
<span dir="ltr" style="unicode-bidi: isolate;">matplotlib</span>.
</strong>
</p>
</div>

```python
import xarray as xr
from matplotlib import pyplot
```

<h2 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
استخدام ملف&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">netCDF4</span>&nbsp;تم تنزيله
</h2>
</div>

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
يمكننا فتح ملف&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">netCDF4</span>&nbsp;
باستخدام الدالة&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">open_dataset()</span>&nbsp;
ضمن مكتبة&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">xarray</span>.
</p>
</div>

```python
ds = xr.open_dataset('data/MERRA2_400.statD_2d_slv_Nx.20231101.nc4')
ds
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
هناك الكثير من العناصر التي يمكن ملاحظتها هنا. تحتوي مجموعة البيانات هذه على ما يلي:
</p>

<ul style="direction: rtl; unicode-bidi: plaintext; text-align: right; padding-right: 1.5em;">
  <li style="direction: rtl; unicode-bidi: plaintext;">
    <strong>الأبعاد (Dimensions):</strong>
    عند العمل مع بيانات خرائطية، تحتوي هذه البيانات على بُعدين على الأقل (مثل خط العرض وخط الطول).
    وإذا كانت الخريطة ناتجة عن بيانات أقمار صناعية تُضيف رصودًا جديدة يوميًا، فيمكن إدخال بُعد ثالث هو الزمن.
    تصف هذه الأبعاد شكل <strong>مكعّب البيانات</strong> الذي يضم ثلاثة <strong>محاور (Axes)</strong>:
    خط الطول (<span dir="ltr" style="unicode-bidi: isolate;">"lon"</span>)،
    وخط العرض (<span dir="ltr" style="unicode-bidi: isolate;">"lat"</span>)،
    والزمن.
  </li>
  <li style="direction: rtl; unicode-bidi: plaintext;">
    <strong>الإحداثيات (Coordinates):</strong>
    تشبه الإحداثيات الأبعاد، لكنها تمثّل المسافات على طول كل محور.
    بالنسبة لخط الطول وخط العرض، تمثّل هذه القيم إحداثيات مركز كل بكسل.
  </li>
  <li style="direction: rtl; unicode-bidi: plaintext;">
    <strong>متغيّرات البيانات (Data Variables):</strong>
    يمكن لملف&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">netCDF4</span>&nbsp;أن يحتوي على عدة متغيّرات مختلفة مُسقطة على الشبكة نفسها.
    على سبيل المثال، قد يحتوي الملف نفسه على كلٍّ من درجة الحرارة الصغرى واليومية العظمى.
  </li>
  <li style="direction: rtl; unicode-bidi: plaintext;">
    <strong>الفهارس (Indexes):</strong>
    تشبه الفهارس الإحداثيات والأبعاد، ولذلك لا نحتاج إلى القلق بشأنها في هذه المرحلة.
  </li>
  <li style="direction: rtl; unicode-bidi: plaintext;">
    <strong>السمات (Attributes):</strong>
    إضافةً إلى قيم البيانات المرسومة على الشبكة، يمكن أن يحتوي ملف&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">netCDF4</span>&nbsp;على
    <strong>بيانات وصفية (Metadata)</strong> تساعد المستخدمين على فهم البيانات.
    تُسجَّل هذه البيانات الوصفية كسمات، وتصف أمورًا مثل إصدار البرمجية المستخدمة لإنشاء البيانات أو الجهة التي أعدّتها.
  </li>
</ul>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
يمكن الوصول إلى المتغيّرات داخل كائن
<span dir="ltr" style="unicode-bidi: isolate;">xarray.Dataset</span>
بطريقة مشابهة للوصول إلى المفاتيح في قاموس&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">Python</span>:
</p>
</div>

```python
ds['T2MMIN']
```

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">

<h2 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
العمل مع كائنات&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">xarray DataArray</span>
</h2>
</div>

```python
tmin = ds['T2MMIN']
tmin.attrs
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
كما هو الحال مع مصفوفات&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">NumPy</span>،
يمكننا التعامل مع كائن&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">DataArray</span>&nbsp;كما لو كان عددًا،
مما يجعل إجراء التحويلات الرياضية على البيانات أمرًا سهلًا.
على سبيل المثال، قد نرغب في تحويل درجات الحرارة الصغرى من كلفن إلى درجات مئوية.
</p>
</div>

```python
# Convert temperatures from deg K to deg C
tmin_c = tmin - 273.15
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
من الأمور التي يجب الانتباه إليها أنه عند إجراء هذا النوع من العمليات،
نفقد السمات المرتبطة بكائن&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">DataArray</span>&nbsp;الأصلي.
ويعود ذلك إلى أن السمات القديمة قد لا تعود مناسبة؛
فمثلًا، نعلم بالفعل أن وحدة القياس القديمة (كلفن) لم تعد صحيحة.
</p>
</div>

```python
tmin_c.attrs
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
يمكننا تعيين السمات في أي وقت باستخدام صيغة القاموس في&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">Python</span>.
</p>
</div>

```python
tmin_c.attrs['units'] = 'degrees C'
tmin_c.attrs
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
ما الفائدة من تعيين سمات جديدة؟
ينبغي القيام بذلك في كل مرة تخطّط فيها لحفظ
<span dir="ltr" style="unicode-bidi: isolate;">Dataset</span>
أو
<span dir="ltr" style="unicode-bidi: isolate;">DataArray</span>
في ملف إخراج ومشاركته مع الآخرين.
توفّر هذه الكائنات الدالة&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">to_netcdf()</span>&nbsp;للقيام بذلك.
</p>
</div>

```python
tmin_c.to_netcdf('example.nc4')
```

ولمزيد من الوضوح حول الأبعاد التي نقوم بالفهرسة عليها،
يمكننا <em>اختيار</em> القيمة/القيم عند إحداثيات معيّنة باستخدام الدالة

```python
ds['T2MMIN'].sel(lat = -90, lon = -180).values
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
ما هي أدنى درجة حرارة في مدينة الجزائر؟
</p>
</div>

```python
ds['T2MMIN'].sel(lat = 36.754, lon = 3.059)
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
ماذا حدث؟
إذا قمنا بفحص إحداثيات مجموعة البيانات، سنلاحظ أنه لا يوجد تطابق دقيق بين خط الطول وخط العرض اللذين قدّمناهما
وقيم إحداثيات مراكز خلايا الشبكة؛ إذ تأتي القيم على فواصل منتظمة قدرها
<span dir="ltr" style="unicode-bidi: isolate;">0.25</span>
أو
<span dir="ltr" style="unicode-bidi: isolate;">0.5</span>
درجة.
</p>
</div>

```python
ds['lon'].values[0:10]
```

```python
ds['lat'].values[0:10]
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
لاحظ أنه يجب علينا تحديد خيار&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">method</span>&nbsp;هنا
لأن إحداثيات مدينة الجزائر لا تتطابق تمامًا مع إحداثيات مراكز خلايا الشبكة؛
أي أننا نطلب استيفاءً باستخدام <strong>أقرب جار (Nearest-neighbor interpolation)</strong>.
</p>
</div>

```python
ds['T2MMIN'].sel(lat = 36.754, lon = 3.059, method = 'nearest').values
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
طريقة أخرى للحصول على النتيجة المطلوبة هي استخدام الدالة
<span dir="ltr" style="unicode-bidi: isolate;">interp()</span>.
<strong>لاحظ أن القيمة الناتجة تختلف قليلًا عن القيمة السابقة.</strong>
ويعود ذلك إلى أن النتيجة السابقة استخدمت الاستيفاء بأقرب جار،
بينما تستخدم الدالة&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">interp()</span>&nbsp;الاستيفاء الخطي افتراضيًا.
وتتوفّر خيارات عديدة أخرى للاستيفاء عبر&nbsp;<span dir="ltr" style="unicode-bidi: isolate;">interp()</span>.
</p>
</div>

```python
ds['T2MMIN'].interp(lat = 36.754, lon = 3.059)
```
