<h2 id="Weather-versus-Climate" style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
الطقس مقابل المناخ
</h2>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
غالبًا ما نسمع كلمتي «الطقس» و«المناخ» تُستخدمان بشكل متبادل. للتوضيح:
</p>

<ul style="direction: rtl; unicode-bidi: plaintext; text-align: right; padding-right: 1.5em;">
  <li style="direction: rtl; unicode-bidi: plaintext;">
    <strong>الطقس</strong>
    يشير إلى الظروف الجوية الحالية، عادةً على مقاييس زمنية قصيرة (ساعية أو يومية).
    على سبيل المثال، قد يشمل طقس «هذا المساء» معدل الهطول المطري بالساعة أو متوسط درجة الحرارة،
    بينما قد يشمل طقس «أمس» إجمالي كمية الأمطار اليومية، أو أعلى درجة حرارة يومية، أو عدد ساعات سطوع الشمس خلال اليوم.
  </li>
  <br>
  <li style="direction: rtl; unicode-bidi: plaintext;">
    <strong>المناخ</strong>
    يشير إلى النمط السائد طويل الأمد للظروف الجوية.
    على سبيل المثال، متوسط كمية الأمطار السنوية يصف مقدار الأمطار الذي يمكنك <em>عادةً</em> توقّعه خلال سنة.
    كما أن توصيف المناخ يعتمد دائمًا على فترة زمنية مرجعية ثابتة، سواء جرى ذكرها صراحةً أم لا.
    فعندما نتحدث عن متوسط درجة الحرارة الشهرية، يجب أن يكون ذلك مبنيًا على فترة مرجعية،
    مثل المتوسط المحسوب عبر السنوات
    <span dir="ltr" style="unicode-bidi: isolate;">1950–2000</span>.
  </li>
</ul>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
<strong>ملاحظة:</strong>
في هذا المنهاج، سنستخدم مصطلح <em>«بيانات المناخ»</em> للإشارة إلى كلا النوعين من البيانات (الطقس والمناخ)، وذلك للتبسيط.
</p>

<h2 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
تباين المناخ في النظم الزراعية
</h2>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
ترتبط دورات الماء والطاقة عبر نظام مناخ الأرض لأن الشمس تسخّن الهواء والسطح،
وتحرّك عملية التمثيل الضوئي، وتؤدي إلى تبخّر المياه.
ولفهم تأثير المناخ على الزراعة، نرغب في قياس المتغيرات المناخية التالية:
</p>

<ul style="direction: rtl; unicode-bidi: plaintext; text-align: right; padding-right: 1.5em;">
  <li style="direction: rtl; unicode-bidi: plaintext;">درجة حرارة الهواء</li>
  <li style="direction: rtl; unicode-bidi: plaintext;">درجة حرارة السطح / التربة</li>
  <li style="direction: rtl; unicode-bidi: plaintext;">الإشعاع الشمسي</li>
  <li style="direction: rtl; unicode-bidi: plaintext;">كمية الهطول المطري</li>
  <li style="direction: rtl; unicode-bidi: plaintext;">التبخر–نتح</li>
  <li style="direction: rtl; unicode-bidi: plaintext;">الرطوبة (أو عجز ضغط البخار)</li>
  <li style="direction: rtl; unicode-bidi: plaintext;">رطوبة التربة</li>
</ul>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
في هذا الجزء من المنهاج، سنناقش حساسات ونماذج وكالة
<span dir="ltr" style="unicode-bidi: isolate;">NASA</span>
التي توفّر هذا النوع من المعلومات على مستوى العالم.
</p>

<h2 id="Issues-with-climate-data-collection" style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
تحديات جمع بيانات المناخ
</h2>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
<strong>من الصعب الحصول على بيانات مناخية تغطي كل مكان وفي كل وقت.</strong>
</p>

<ul style="direction: rtl; unicode-bidi: plaintext; text-align: right; padding-right: 1.5em;">
  <li style="direction: rtl; unicode-bidi: plaintext;">
    لا يمكن وضع محطة أرصاد جوية في كل نقطة على سطح الأرض.
    كما أننا لا نعرف على وجه اليقين ما هي المسافة المثلى بين المحطات؛ إذ قد تختلف حسب التطبيق أو المنطقة.
  </li>
  <br>
  <li style="direction: rtl; unicode-bidi: plaintext;">
    تمتلك محطات الأرصاد تغطية مكانية محدودة.
    فدرجة الحرارة والرطوبة، على سبيل المثال، قد تتغيران على مسافات قصيرة جدًا (قد تكون بضعة أمتار فقط)،
    لذا قد تمثل محطة واحدة مساحة صغيرة فحسب.
  </li>
  <br>
  <li style="direction: rtl; unicode-bidi: plaintext;">
    في بعض مناطق العالم، تُحجب أنظمة الرادار الجوي بواسطة التلال أو الجبال، ما يحدّ من نطاق تغطيتها.
  </li>
</ul>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
وبسبب القيود المذكورة أعلاه، إضافةً إلى القيود الحاسوبية، فإن البيانات الناتجة عن نماذج الطقس تكون ذات دقة مكانية محدودة.
وأحيانًا تصف النماذج درجة الحرارة أو سرعة الرياح لمساحة أكبر بكثير من حقل مزارع، أو حتى أكبر من مدينة أو وحدة إدارية.
</p>

<h2 id="Sensors-and-ground-stations" style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
الحساسات ومحطات الرصد الأرضية
</h2>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
توجد مصادر عديدة لبيانات الطقس.
فبالإضافة إلى الحساسات المثبتة في محطات الرصد الأرضية، توجد حساسات جوية وأقمار صناعية تدور حول الأرض وتوفّر بيانات آنية عن الظروف الجوية.
<strong>تأمل مصادر بيانات الطقس (والمناخ) المختلفة في الشكل أدناه.</strong>
</p>

</p>

</p>

<h2 id="Interpolating-station-data" style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
استيفاء بيانات المحطات
</h2>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
يمكن تقدير الظروف الجوية في موقع يصعب الوصول إليه إذا توفرت محطة رصد قريبة (محطة مرجعية).
هناك عدة مبادئ فيزيائية يمكن استخدامها لاستيفاء الظروف الجوية اعتمادًا على موقع واحد أو أكثر من المواقع القريبة التي تُقاس فيها عناصر الطقس.
<strong>نعرف مثلًا أن درجة حرارة الهواء تنخفض مع الارتفاع بمعدل معين يعتمد على الرطوبة (انظر الشكل أدناه).</strong>
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
كما نعرف أنه في التضاريس الجبلية، تبقى بعض المناطق في الظل لفترات أطول من غيرها.
وقد يؤثر الغطاء النباتي في «درجة الحرارة المحسوسة»، لأن النبات يظلل سطح الأرض،
ولأن الغطاء النباتي نفسه قد يكون مصدرًا بسيطًا للحرارة في الطقس البارد.
تساعد هذه العوامل وغيرها على استنتاج ظروف الطقس في مواقع غير مراقبة باستخدام نموذج
قد يكون نموذجًا إحصائيًا بسيطًا أو نموذجًا فيزيائيًا معقدًا.
</p>
