<h2 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
كيفية استخدام Earthdata Search
</h2>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
لنتعرّف على بوابة
<span dir="ltr" style="unicode-bidi: isolate;">Earthdata Search</span>
من خلال البحث عن بعض بيانات المناخ.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
بعد تسجيل الدخول، انظر إلى اللوحة الجانبية الموجودة على الجهة اليسرى من الواجهة.
ضمن قسم
<span dir="ltr" style="unicode-bidi: isolate;">"Filter Collections"</span>،
قم بتوسيع العنوان الذي يحمل اسم
<span dir="ltr" style="unicode-bidi: isolate;">"Keywords"</span>.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
يوضّح الشكل أدناه مثالًا على ذلك.
</p>

<img src="./assets/M1_screenshot_Earthdata_Search_keywords.png" alt="Earthdata Search keywords screenshot">

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
بعد ذلك، قم بتحديد خيار
<span dir="ltr" style="unicode-bidi: isolate;">"Climate Indicators"</span>.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
قد تبدو الخيارات المعروضة في هذا القسم كثيرة ومعقّدة في البداية.
لذا، دعونا نفكّك المعلومات المعروضة في الجهة اليمنى من لوحة البحث خطوةً بخطوة.
</p>

<ul style="direction: rtl; unicode-bidi: plaintext; text-align: right; padding-right: 1.5em;">
  <li style="direction: rtl; unicode-bidi: plaintext;">
    تحتوي كل مجموعة بيانات على عدد معيّن من
    <strong>الملفّات البيانية (Granules)</strong>،
    وهو مصطلح يُستخدم للإشارة إلى ملف بيانات واحد.
  </li>

  <li style="direction: rtl; unicode-bidi: plaintext;">
    تمتد كل مجموعة بيانات على فترة زمنية محددة؛
    فعلى سبيل المثال، تشير العبارة
    <span dir="ltr" style="unicode-bidi: isolate;">2003-02-03 ongoing</span>
    إلى أن البيانات جُمعت بشكل مستمر منذ
    <span dir="ltr" style="unicode-bidi: isolate;">3 فبراير 2003</span>.
  </li>

  <li style="direction: rtl; unicode-bidi: plaintext;">
    تحتوي بعض مجموعات البيانات على أيقونة
    <span dir="ltr" style="unicode-bidi: isolate;">"Earthdata Cloud"</span>،
    ما يعني أن البيانات متاحة عبر السحابة دون الحاجة إلى تنزيل الملفات على جهازك.
    سنتعرّف على كيفية الاستفادة من ذلك لاحقًا في هذا المنهاج.
  </li>
</ul>

</div>


<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">

<h3 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
معلومات حول كل مجموعة بيانات
</h3>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
عند تمرير مؤشر الفأرة فوق إحدى مجموعات البيانات في الجهة اليمنى من هذه اللوحة،
ستظهر أيقونة معلومات
<span dir="ltr" style="unicode-bidi: isolate;">(i)</span>.
يوضّح الشكل أدناه مثالًا على ذلك.
يرجى النقر على زر
<span dir="ltr" style="unicode-bidi: isolate;">(i)</span>.
</p>

<img src="./assets/M1_screenshot_Earthdata_Search_dataset_hovering.png"
     alt="Earthdata Search dataset hovering screenshot">

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
تتوفّر في هذه الصفحة كمية كبيرة من المعلومات المهمة.
في أعلى صفحة المعلومات، ستلاحظ وجود ثلاثة مربّعات، كما هو موضّح في المثال التالي.
</p>

<img src="./assets/M1_screenshot_Earthdata_Search_info_page.png"
     alt="Earthdata Search dataset information page screenshot">

<ul style="direction: rtl; unicode-bidi: plaintext; text-align: right; padding-right: 1.5em;">

  <li style="direction: rtl; unicode-bidi: plaintext;">
    <strong>اسم مختصر يعرّف مجموعة البيانات بشكل فريد</strong>؛
    في هذا المثال هو:
    <span dir="ltr" style="unicode-bidi: isolate;">
      GRACEADM_CLSM025GL_7D
    </span>
  </li>

  <li style="direction: rtl; unicode-bidi: plaintext;">
    <strong>رقم الإصدار</strong>؛
    في هذا المثال هو:
    <span dir="ltr" style="unicode-bidi: isolate;">
      Version&nbsp;3.0
    </span>
  </li>

  <li style="direction: rtl; unicode-bidi: plaintext;">
    <strong>المعرّف الرقمي للكائنات (DOI)</strong>.
    يشبه الـ
    <span dir="ltr" style="unicode-bidi: isolate;">DOI</span>
    عنوان الموقع الإلكتروني، لكنه يوفّر معرّفًا فريدًا لأي نوع من المعلومات الرقمية،
    وغالبًا ما يُستخدم لتعريف مجموعات البيانات.
    عند استخدام أي مجموعة بيانات في بحثك العلمي،
    يجب التأكّد من الاستشهاد بمعرّف
    <span dir="ltr" style="unicode-bidi: isolate;">DOI</span>
    الخاص بها في أي منشور علمي.
  </li>

</ul>

<h2 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
البحث عن البيانات
</h2>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
إضافةً إلى تصفّح مجموعات البيانات حسب المجال الموضوعي
(مثل
<span dir="ltr" style="unicode-bidi: isolate;">Climate Indicators</span>)،
يمكننا أيضًا البحث عن مجموعات البيانات باستخدام الاسم أو الكلمات المفتاحية،
تمامًا كما نفعل عند استخدام محرّكات البحث على الإنترنت.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
اكتب
<span dir="ltr" style="unicode-bidi: isolate;">MERRA-2</span>
في مربّع البحث الموجود في أعلى يسار الصفحة، كما هو موضّح في لقطة الشاشة أدناه.
</p>

<img src="./assets/M1_screenshot_Earthdata_Search_text_box_search.png"
     alt="Earthdata Search text box search screenshot">

<h2 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
مجموعات بيانات إعادة التحليل
</h2>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
تستخدم
<strong>مجموعات بيانات إعادة التحليل</strong>
تقنيات
<span dir="ltr" style="unicode-bidi: isolate;">استيعاب البيانات (Data Assimilation)</span>
لدمج بيانات الرصد الأرضي التاريخية مع نماذج التنبؤ العددي بالطقس.
في هذه العملية، يحاول النموذج أولًا التنبؤ بحالة جوية معيّنة
(على سبيل المثال، معدل الهطول المطري في
<span dir="ltr" style="unicode-bidi: isolate;">31 أغسطس 2020</span>)،
ثم تُقارَن القيمة المتوقعة بالقيمة الفعلية التي تم رصدها.
إذا وُجد اختلاف كبير بين القيمتين، يتم تعديل النموذج لتحسين قدرته على التنبؤ.
</p>

<img src="./assets/M1_fig_re-analysis.png"
     alt="Re-analysis concept diagram">

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
تُعد مجموعة بيانات
<a href="https://gmao.gsfc.nasa.gov/reanalysis/MERRA-2/" dir="ltr">
Modern-Era Retrospective Analysis for Research and Applications (MERRA-2)
</a>
مجموعة إعادة التحليل الرئيسية التابعة لوكالة
<span dir="ltr" style="unicode-bidi: isolate;">NASA</span>.
وهي تدمج بيانات قادمة من عدة منصّات أقمار صناعية ومحطات أرصاد جوية أرضية
من أجل التنبؤ بالظروف المناخية على مستوى العالم.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
توفّر
<span dir="ltr" style="unicode-bidi: isolate;">MERRA-2</span>
عددًا كبيرًا من المتغيّرات الجوية المختلفة ضمن شبكة عالمية متّسقة،
بدقة مكانية تبلغ
<span dir="ltr" style="unicode-bidi: isolate;">0.625</span>
درجة طولية ×
<span dir="ltr" style="unicode-bidi: isolate;">0.5</span>
درجة عرضية.
وكما هو الحال في نماذج الطقس الأخرى، تحتوي البيانات على عدة طبقات رأسية،
تمثّل إمّا الارتفاع النسبي عن سطح الأرض أو مستويات الضغط في الغلاف الجوي.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
نظرًا للوقت المطلوب لاستيعاب ومعالجة هذا الكم الكبير من البيانات،
تتأخر بيانات
<span dir="ltr" style="unicode-bidi: isolate;">MERRA-2</span>
دائمًا بضعة أسابيع عن الزمن الحقيقي.
لذلك، وكما هو الحال مع جميع مجموعات بيانات إعادة التحليل،
ينبغي استخدام هذه البيانات فقط لوصف الظروف المناخية السائدة أو التاريخية،
وليس الظروف الآنية.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
تنتج
<span dir="ltr" style="unicode-bidi: isolate;">MERRA-2</span>
نوعين رئيسيين من المنتجات:
<strong>منتجات آنية (Instantaneous)</strong>
و
<strong>منتجات متوسّطة زمنيًا (Time-averaged)</strong>.
تمثّل المنتجات الآنية لقطات لحظية للظروف الجوية.
فعلى سبيل المثال، إذا كانت سرعة الرياح المتجهة شمالًا في منتج آني
ذو فترة
<span dir="ltr" style="unicode-bidi: isolate;">3 ساعات</span>
تساوي
<span dir="ltr" style="unicode-bidi: isolate;">2</span>
متر في الثانية،
فإن هذه القيمة تمثّل سرعة الرياح التي حاكاها النموذج في تلك اللحظة الزمنية الدقيقة،
مثل
<span dir="ltr" style="unicode-bidi: isolate;">15:00 UTC</span>.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
أما في المنتجات المتوسّطة زمنيًا ذات الفترة
<span dir="ltr" style="unicode-bidi: isolate;">3 ساعات</span>،
فإن قيمة سرعة الرياح عند
<span dir="ltr" style="unicode-bidi: isolate;">13:30 UTC</span>
تمثّل المتوسّط المحسوب خلال الثلاث ساعات المحيطة بتلك اللحظة،
أي من
<span dir="ltr" style="unicode-bidi: isolate;">12:00 UTC</span>
إلى
<span dir="ltr" style="unicode-bidi: isolate;">15:00 UTC</span>.
كما تتوفّر أيضًا منتجات بمتوسطات يومية،
مثل درجات الحرارة الصغرى والعظمى اليومية.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
وبالمثل، يمكن أن تكون منتجات بيانات
<span dir="ltr" style="unicode-bidi: isolate;">MERRA-2</span>
إمّا
<strong>ثنائية الأبعاد (2D)</strong>
أو
<strong>ثلاثية الأبعاد (3D)</strong>.
في الحالة ثلاثية الأبعاد، توجد عدة شبكات عرض–طول لمستويات رأسية مختلفة،
غالبًا ما تُعبّر عنها بمستويات ضغط جوي مختلفة.
أما علماء الزراعة والمزارعون ومديرو الأراضي،
فغالبًا ما يكون اهتمامهم أكبر بالبيانات ثنائية الأبعاد،
والتي تُعرف أيضًا باسم بيانات
<strong>المستوى الواحد (Single-level – slv)</strong>.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
عادةً ما تُوسم منتجات
<span dir="ltr" style="unicode-bidi: isolate;">MERRA-2</span>
من هذا النوع بالشكل التالي:
</p>

<pre dir="ltr">
inst1_2d_slv_*
inst3_2d_slv_*
</pre>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
تشير البادئة
<span dir="ltr" style="unicode-bidi: isolate;">inst</span>
إلى بيانات آنية بفواصل زمنية قدرها ساعة واحدة أو ثلاث ساعات.
وقد تجد بدلًا من ذلك منتجات متوسّطة زمنيًا موسومة بالشكل التالي:
</p>

<pre dir="ltr">
tavg1_2d_slv_*
tavg3_2d_slv_*
</pre>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
حيث تشير
<span dir="ltr" style="unicode-bidi: isolate;">tavg1</span>
و
<span dir="ltr" style="unicode-bidi: isolate;">tavg3</span>
إلى بيانات متوسّطة زمنيًا بفواصل ساعة واحدة أو ثلاث ساعات، على التوالي.
وبدلًا من
<span dir="ltr" style="unicode-bidi: isolate;">slv</span>،
قد تجد منتجات موسومة بـ
<span dir="ltr" style="unicode-bidi: isolate;">rad</span>
لبيانات الإشعاع الشمسي والحراري،
أو
<span dir="ltr" style="unicode-bidi: isolate;">aer</span>
لبيانات الهباء الجوي.
وتتوفّر من هذه المنتجات نسخ ثنائية وثلاثية الأبعاد.
</p>

</div>

<h2 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
بيانات MERRA-2 في Earthdata Search
</h2>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
لنلقِ نظرة على بعض مجموعات بيانات
<span dir="ltr" style="unicode-bidi: isolate;">MERRA-2</span>.
لنفترض أننا نرغب في تحديد درجة الحرارة الصغرى اليومية ضمن منطقة اهتمامنا.
من خلال توثيق
<span dir="ltr" style="unicode-bidi: isolate;">MERRA-2</span>،
نتعرّف على أن مجموعة البيانات المطلوبة تحمل الاسم:
<span dir="ltr" style="unicode-bidi: isolate;">statD_2d_slv_Nx</span>.
</p>

<ol style="direction: rtl; unicode-bidi: plaintext; text-align: right; padding-right: 1.5em;">
  <li style="direction: rtl; unicode-bidi: plaintext;">
    اكتب
    <span dir="ltr" style="unicode-bidi: isolate;">statD_2d_slv_Nx</span>
    في مربّع البحث الموجود في أعلى يسار واجهة
    <span dir="ltr" style="unicode-bidi: isolate;">Earthdata Search</span>.
  </li>

  <li style="direction: rtl; unicode-bidi: plaintext;">
    انقر على المربّع الذي يظهر في نتائج البحث
    (ينبغي أن تكون هناك نتيجة واحدة فقط).
  </li>

  <li style="direction: rtl; unicode-bidi: plaintext;">
    انتقل إلى الجهة اليسرى من الواجهة، وتحديدًا إلى قسم
    <span dir="ltr" style="unicode-bidi: isolate;">"Filter Granules"</span>.
    ستلاحظ أن خيار التصفية الوحيد المتاح هنا هو التصفية الزمنية.
    ويعود ذلك إلى أن كل ملف بيانات من
    <span dir="ltr" style="unicode-bidi: isolate;">MERRA-2</span>
    يغطي الكرة الأرضية كاملة، وبالتالي لا تكون التصفية المكانية ضرورية.
    عند إدخال تاريخ في خانة
    <span dir="ltr" style="unicode-bidi: isolate;">"Start"</span>،
    ستظهر فقط الملفات اللاحقة لذلك التاريخ.
    أما عند إدخال تاريخ في خانيتي
    <span dir="ltr" style="unicode-bidi: isolate;">"Start"</span>
    و
    <span dir="ltr" style="unicode-bidi: isolate;">"End"</span>،
    فستظهر فقط الملفات الواقعة بين هذين التاريخين.
  </li>

  <li style="direction: rtl; unicode-bidi: plaintext;">
    في حال النقر على زر التنزيل في هذه المرحلة،
    سيتم تنزيل ملف واحد ينتهي بامتداد
    <span dir="ltr" style="unicode-bidi: isolate;">*.nc4</span>.
    يرجى النقر لتنزيل أي ملف من الملفات المعروضة،
    كما هو موضّح في لقطة الشاشة أدناه.
  </li>
</ol>
