<h2 id="General-Circulation-Models" style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
نماذج الدوران العامة
</h2>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
تُعد نماذج الدوران العامة
(<span dir="ltr">General Circulation Models – GCMs</span>)
تمثيلات رياضية معقّدة لعمليات تبادل الإشعاع والحرارة والغازات النزرة في الغلاف الجوي.
كما يظهر في المخطط أدناه، تقوم هذه النماذج بمحاكاة المحيطات والغلاف الجوي وسطح اليابسة
على هيئة مجموعة من الخلايا الشبكية المنفصلة.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
يمكن لكل خلية شبكية أن تتبادل الكتلة أو الطاقة مع الخلايا المجاورة لها.
وكلما صغر حجم الخلايا الشبكية، تحسّن تمثيل الواقع الفيزيائي للأرض،
لكن ذلك يؤدي في المقابل إلى زيادة كبيرة في التعقيد الحسابي.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
في نماذج الدوران العامة، يتم تمثيل الغلاف الجوي ثلاثي الأبعاد بالكامل
وتشغيله زمنيًا لمحاكاة التغيّرات المناخية في الماضي والحاضر والمستقبل.
ويمكن لهذه النماذج التنبؤ بالأحداث الجوية على مقاييس زمنية تمتد من ساعات إلى أيام،
وأحيانًا لفترات أطول، مع العلم أن بعض الظواهر لا يمكن التنبؤ بها بدقة
إلا على مدد زمنية قصيرة.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
تُعد نماذج الدوران العامة الأداة الأساسية لنمذجة التغيّر المناخي.
وتشير مقارنة تنبؤات هذه النماذج مع المناخ التاريخي،
وخاصة اتجاهات وشذوذات درجات الحرارة،
إلى أن نماذج الدوران العامة قادرة على تمثيل النظام المناخي للأرض
بدقة معقولة.
إلا أن دقتها المكانية تبقى محدودة نسبيًا،
حيث تكون الخلايا الشبكية عادةً في حدود
<span dir="ltr">100</span>
كيلومتر مربع.
</p>

<h2 id="Earth-System-Models" style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
نماذج نظام الأرض
(<span dir="ltr">Earth System Models – ESMs</span>)
</h2>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
تمثّل نماذج الدوران العامة تبادل المياه والحرارة والغازات
بين سطح اليابسة والمحيطات والغلاف الجوي،
إلا أنها لا تتضمّن تمثيلًا مباشرًا للمجتمعات البشرية
وكيفية تأثيرها في هذه العمليات.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
على النقيض من ذلك، فإن
<strong>نماذج نظام الأرض</strong>
قادرة على تمثيل جميع العمليات الفيزيائية الموجودة في نماذج الدوران العامة،
إضافةً إلى عمليات بشرية وبيئية مثل إزالة الغابات،
والزراعة، وتخزين أو إطلاق الكربون من التربة والغابات.
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
يوضح الشكل أدناه (من
<span dir="ltr">Bonan &amp; Doney (2018, Science)</span>)
بعض الفروقات الرئيسية بين نماذج الدوران العامة ونماذج نظام الأرض.
</p>

<img src="./assets/M1_fig_Bonan_Doney_2018.jpg">

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
<em>الصورة من Bonan &amp; Doney (2018)، مجلة Science</em>
</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">
ونظرًا لتضمينها التغيّرات التي يسببها الإنسان في النظم البيئية،
يمكن استخدام نماذج نظام الأرض في
<strong>نمذجة السيناريوهات</strong>
لدراسة التغيّرات المستقبلية المحتملة.
</p>
