## الطقس مقابل المناخ

غالبًا ما نسمع كلمتي "الطقس" و"المناخ" تُستخدمان بالتبادل. للتوضيح:

- **الطقس** يشير إلى الظروف الجوية الحالية، عادةً على مقاييس زمنية بالساعة أو اليوم. على سبيل المثال، الطقس "هذا المساء" قد يشمل معدل هطول الأمطار بالساعة أو متوسط درجة الحرارة. الطقس "أمس" قد يشمل كمية الأمطار اليومية الإجمالية، أو الحد الأقصى لدرجة الحرارة اليومية، أو عدد ساعات سطوع الشمس اليومية.

- **المناخ** يشير إلى النمط السائد طويل الأمد للظروف الجوية. على سبيل المثال، تصف كمية الأمطار السنوية المتوسطة مقدار الأمطار التي يمكنك *توقعها عادةً* في السنة. وصف المناخ يعتمد دائمًا على فترة زمنية ثابتة، سواء تم تحديدها أم لا. عندما نتحدث عن متوسط درجة الحرارة الشهرية، على سبيل المثال، يجب أن يكون ذلك استنادًا إلى فترة مرجعية، مثل المتوسط على مدار السنوات 1950 إلى 2000.

**ومع ذلك، في هذا المنهاج، غالبًا ما نستخدم مصطلح *"بيانات المناخ"* للإشارة إلى كلا النوعين من البيانات، لتبسيط الأمر.**

## تباين المناخ في الأنظمة الزراعية

يرتبط دوران الماء والطاقة في نظام المناخ الأرضي لأن الشمس تدفئ هواء وسطح الأرض، وتحفز التمثيل الضوئي، وتبخر المياه. لفهم تأثير المناخ على الزراعة، نود قياس المتغيرات المناخية التالية:

- درجة حرارة الهواء
- درجة حرارة السطح/التربة
- الإشعاع الشمسي
- كمية الهطول
- التبخر والنتح
- الرطوبة (أو عجز ضغط البخار)
- رطوبة التربة

في هذه الوحدة، سنناقش أجهزة استشعار ونماذج وكالة ناسا التي توفر هذا النوع من المعلومات على مستوى العالم.

## مشاكل جمع بيانات المناخ

**من الصعب الحصول على بيانات المناخ لكل جزء من العالم وفي كل وقت.**

- لا يمكننا وضع محطة طقس في كل نقطة على الأرض. حتى أننا لا نعرف بالضبط مدى قرب محطات الطقس من بعضها البعض، فقد يختلف ذلك بناءً على التطبيقات المختلفة أو المناطق المختلفة من العالم.

- وبالمثل، فإن محطات الطقس لها تغطية محدودة. يمكن أن تختلف درجة الحرارة والرطوبة، على سبيل المثال، على مسافات قصيرة تصل إلى بضعة أمتار. لذلك قد تمثل أي محطة طقس منطقة صغيرة فقط.

- في بعض مناطق العالم، تعيق التلال أو الجبال أنظمة الرادار الجوية، مما يحد من تغطيتها.

نظرًا للقيود المذكورة أعلاه وكذلك القيود الحسابية، فإن البيانات التي تنتجها نماذج الطقس لها دقة مكانية محدودة. أحيانًا، قد يصف النموذج درجة الحرارة أو سرعة الرياح لمنطقة أكبر بكثير من حقل مزارع أو حتى أكبر من بلدة أو مقاطعة إدارية.
