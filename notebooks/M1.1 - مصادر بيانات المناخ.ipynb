## الطقس مقابل المناخ

غالبًا ما نسمع كلمات "الطقس" و"المناخ" تُستخدم بالتبادل. للتوضيح:

- **الطقس** يشير إلى الظروف الجوية الحالية، عادة على فترات زمنية قصيرة مثل الساعات أو الأيام. على سبيل المثال، قد يتضمن الطقس "لهذا اليوم" معدل هطول الأمطار في الساعة أو متوسط درجة الحرارة. أما الطقس "أمس" فقد يشمل إجمالي كمية الأمطار اليومية أو أقصى درجة حرارة يومية أو عدد ساعات سطوع الشمس اليومية.

- **المناخ** يشير إلى النمط السائد للظروف الجوية على المدى الطويل. على سبيل المثال، يشير متوسط كمية هطول الأمطار السنوية إلى مقدار الأمطار الذي يمكن أن *تتوقعه عادة* في السنة. تستند أوصاف المناخ دائمًا إلى فترة زمنية ثابتة، سواء كانت محددة أو غير محددة. عندما نتحدث عن متوسط درجة الحرارة الشهرية، على سبيل المثال، يجب أن يستند ذلك إلى فترة مرجعية، مثل المتوسط على مدار سنوات 1950 إلى 2000.

**ومع ذلك، في هذا المنهج، سنستخدم غالبًا مصطلح *"بيانات المناخ"* للإشارة إلى كلا النوعين من البيانات، للتبسيط.**


## تباين المناخ في النظم الزراعية

يرتبط دوران الماء والطاقة في نظام مناخ الأرض لأن الشمس تسخن الهواء وسطح الأرض، وتحفز عملية التمثيل الضوئي، وتؤدي إلى تبخر المياه. لفهم تأثيرات المناخ على الزراعة، نرغب في قياس المتغيرات المناخية التالية:

- درجة حرارة الهواء
- درجة حرارة السطح/ التربة
- الإشعاع الشمسي
- كمية الهطول
- النتح التبخري
- الرطوبة (أو نقص ضغط البخار)
- رطوبة التربة

في هذه الوحدة، سنتحدث عن أجهزة الاستشعار والنماذج التابعة لوكالة ناسا التي توفر هذا النوع من المعلومات على مستوى العالم.


## مشاكل جمع بيانات المناخ

**من الصعب الحصول على بيانات المناخ لكل جزء من العالم في كل الأوقات.**

- لا يمكننا وضع محطة طقس في كل نقطة على الأرض. ولا نعرف حتى على وجه التحديد مدى قرب المحطات التي يجب أن تكون - فقد يختلف ذلك باختلاف التطبيقات أو مناطق العالم المختلفة.

- وبالمثل، فإن محطات الطقس لها تغطية محدودة. على سبيل المثال، يمكن أن تتغير درجة الحرارة والرطوبة على مسافات قصيرة تصل إلى بضعة أمتار. لذا قد تمثل أي محطة طقس منطقة صغيرة فقط.

- في بعض مناطق العالم، تكون أنظمة رادار الطقس محجوبة بسبب التلال أو الجبال، مما يحد من تغطيتها.

بسبب القيود المذكورة أعلاه وأيضًا القيود الحسابية، فإن البيانات التي تنتجها نماذج الطقس لها دقة مكانية محدودة. أحيانًا قد يصف النموذج درجة الحرارة أو سرعة الرياح لمنطقة أكبر بكثير من حقل المزارع أو حتى أكبر من مدينة أو منطقة إدارية.


## أجهزة الاستشعار والمحطات الأرضية

هناك العديد من المصادر المختلفة لبيانات الطقس. بالإضافة إلى أجهزة الاستشعار المثبتة على محطات الطقس الأرضية، هناك العديد من أجهزة الاستشعار المحمولة جوًا والأقمار الصناعية التي تدور حول الأرض والتي توفر بيانات في الوقت الحقيقي حول الظروف الجوية. **فكر في المصادر المختلفة لبيانات الطقس (والمناخ) في الصورة أدناه.**

## استقراء بيانات المحطات

من الممكن تحديد الظروف الجوية في مكان يصعب الوصول إليه، بشرط أن تكون هناك محطة أساسية قريبة. هناك عدة مبادئ فيزيائية يمكننا استخدامها لاستقراء الظروف الجوية في موقع بناءً على موقع أو أكثر قريب حيث يتم مراقبة الطقس. **نحن نعلم أن درجة حرارة الهواء تنخفض مع الارتفاع بمعدل معين بناءً على الرطوبة (انظر الشكل أدناه).** 

نعلم أيضًا أنه في التضاريس الجبلية، تكون بعض المناطق مظللة لفترات أطول من غيرها. يمكن أن يؤثر الغطاء النباتي على درجة الحرارة المحسوسة لأن النبات يظلّل الأرض وفي الطقس البارد، قد يكون مصدرًا صغيرًا للحرارة. يمكن أن تساعدنا هذه العوامل الفيزيائية وغيرها في استنتاج الظروف الجوية في موقع غير مراقب باستخدام نموذج. قد يكون هذا النموذج بسيطًا إحصائيًا أو معقدًا فيزيائيًا.
